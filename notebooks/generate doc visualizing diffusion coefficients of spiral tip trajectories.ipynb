# define module for generating pdf from png's found

```python
#combine two images
def concat_horz(im1, im2):
    dst = Image.new('RGB', (im1.width + im2.width, im1.height))
    dst.paste(im1, (0, 0))
    dst.paste(im2, (im1.width, 0))
    return dst
def concat_vert(im1, im2):
    dst = Image.new('RGB', (im1.width, im1.height + im2.height))
    dst.paste(im1, (0, 0))
    dst.paste(im2, (0, im1.height))
    return dst
# Example Usage: get_concat_h(im1, im1).save('data/dst/pillow_concat_h.jpg')

def vert_concat_png_list(fn_lst):
    '''fn_lst is a list of directories to image (png) files'''
    image = Image.open(fn_lst.pop())
    im_out= image.convert('RGB')
    for fn in fn_lst:
        image = Image.open(fn)
        im_out= concat_vert(im_out,image.convert('RGB'))
    return im_out
```

```python
def get_msd_png_files(trial_folder_name, ic_suite_dir):
    os.chdir(os.path.join(ic_suite_dir,trial_folder_name))
    def is_png(fn): return '.png'==fn[-4:]
    if not os.path.exists('msd'):
        print(f"trial not yet analyzed: {trial_folder_name}")        
        return False
    os.chdir('msd')
    fn_lst=sorted((os.path.abspath(fn) for fn in os.listdir() if is_png(fn)))
    if os.path.exists('fig'):
        os.chdir('fig')
        fn_lst2=sorted((os.path.abspath(fn) for fn in os.listdir() if is_png(fn)))
        fn_lst.extend(fn_lst2)
    if len(fn_lst)==0:
        print(f"no figures found for {trial_folder_name}")
    return fn_lst

def doc_gener(trial_folder_name, ic_suite_dir):
    fn_lst=get_msd_png_files(trial_folder_name, ic_suite_dir)
    if not fn_lst:
        return False
    save_fn=f'diffusive_{trial_folder_name}.png'
    save_folder=ic_suite_dir[:ic_suite_dir.find('care/notebooks')+len('care/')]+'doc/diffusion'
    im_out = vert_concat_png_list(fn_lst)
    #save out Image as pdf
    if not os.path.exists(save_folder):
        os.mkdir(save_folder)
    os.chdir(save_folder)
    im_out.save(save_fn)
    print(f'doc_out saved in care/doc/diffusion for {trial_folder_name}/')
    return True
```
